```python
def trapezoidal(x,a,b,c,d):
    if(x<a or x>d):
        return 0
    else:
        if(a<=x<b):
            return ((x-a)/(b-a))
        else:
            if(b<=x<c):
                return 1
            else:
                if (c<=x<=d):
                    return ((d-x)/(d-c))
```

```python
def fuzzify(element):
    if element<0.5:
        return round(trapezoidal(element,0,.25,.5,.75),2)
    else:
        return round(trapezoidal(element,0.25,.5,.75,1),2)
def fuzz(element):
    if element<0.5:
        return "Low"
    else:
        return "High"
def fuz_diab(element):
    if element==1:
        return "diabetic"
    else:
        return "Non-diabetic"
fuzzify=np.vectorize(fuzzify)
fuz_diab=np.vectorize(fuz_diab)
fuzz=np.vectorize(fuzz)
```

```python
def main(X_train):
    fuzzy_df=np.zeros(X_train.shape)
    for i in range(X_train.shape[0]):
        fuzzy_df[i]=fuzzify(X_train.iloc[i,:])
    return fuzzy_df
```

```python
main(X_train)
```

```python
def ling(fuzzy_df,y_train):
    Class=fuz_diab(y_train)
    fuzzy_df=fuzz(fuzzy_df)
    new_df=np.column_stack((fuzzy_df,Class))
    return new_df
```

# Blended Crossover operator

```python
import random
def bcov(mat):#mat=fuzzy_df
    k,j=np.random.choice([x for x in range(1,len(mat))],2)
    x1,x2=mat.iloc[k,:],mat.iloc[j,:]
    alpha=random.uniform(0,1)
    r=random.uniform(0,1)
    a=np.subtract(x1,alpha*(x2-x1))
    b=np.subtract(x2,alpha*(x2-x1))
    Y_val=a+r*(b-a)
    if any(np.greater_equal(Y_val,min(x1))) and any(np.less_equal(Y_val,max(x2))):
        return Y_val
    else:
        bcov(mat)
```
